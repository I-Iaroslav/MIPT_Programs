```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('out.txt', sep=" ", header=None)
##print(df)
Grid = np.asmatrix(df.values)[:,0:-1]
print(Grid)
```

```python
fig, ax = plt.subplots(figsize = (9, 7))
extent = 0, 1, 0, 1
cb = ax.imshow(Grid, cmap=plt.cm.jet, interpolation="spline16", origin='lower', extent=extent)
fig.colorbar(cb)
fig.tight_layout()
fig.savefig("Solution.pdf")
plt.show()
```
