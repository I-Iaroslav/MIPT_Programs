```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
buff = []
with open("fft.txt") as f:
    for line in f:
        buff.append([float(x) for x in line.split()])
data = buff[0]
x = [i for i in range(0, len(data))]
```

```python
plt.figure(num=1, figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')
plt.title("FFT") # заголовок
plt.xlabel("x") # ось абсцисс
plt.ylabel("y(x)") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(x, data, '-', label="func")
plt.legend()
plt.show()
```
